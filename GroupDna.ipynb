**READ THE CHAT FILE**

```python
file_name = "hostel_bois.txt"

with open(file_name, "r", encoding="utf-8") as file:
    chat = file.readlines()

print("Total lines:", len(chat))
```

**PARSE THE COMPLETE CHAT FILE**

```python
messages = []

for line in chat:
    line = line.strip()

    if not line:
        continue

    if " - " not in line:
        continue

    parts = line.split(" - ", 1)

    date_time = parts[0]
    content = parts[1]

    if ", " not in date_time:
        continue

    date, time = date_time.split(", ", 1)

    if ": " not in content:
        continue

    sender, message = content.split(": ", 1)

    entry = {
        "date": date,
        "time": time,
        "sender": sender,
        "message": message
    }

    messages.append(entry)

print("Total Parsed Messages:", len(messages))
```
